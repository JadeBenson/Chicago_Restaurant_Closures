```python
import numpy as np
import pandas as pd
import csv
```

```python
food_inspect = pd.read_csv("Food_Inspections.csv")
```

```python
#what's the size of the df? 
len(food_inspect)
```

```python
food_inspect.head()
```

```python
#convert inspection date to datetime object 
#year-month-day
food_inspect['Inspection Date'] = pd.to_datetime(food_inspect['Inspection Date'])
food_inspect.head()
```

```python
#all inspections since Jan 1, 2019
inspect_subset = food_inspect[food_inspect['Inspection Date'] > '2019-01-01']
```

```python
#cut down so from jan 1, 2019 - march 1, 2020 
inspect_subset = inspect_subset[inspect_subset["Inspection Date"] < "2020-03-01"]
len(inspect_subset)
```

```python
#sort it by date
inspect_subset = inspect_subset.sort_values(by = ["Inspection Date"])
```
